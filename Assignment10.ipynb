Q1. Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports, technology, food, books, etc.).**

1. Convert text to lowercase and remove punctuation using re.  
2. Tokenize the text into words and sentences.  
3. Split using split() and word_tokenize() and compare how Python split and NLTK’s word_tokenize() differ.  
4. Remove stopwords (using NLTK's stopwords list).  
5. Display word frequency distribution (excluding stopwords).


```python
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

paragraph = """Russia’s fight against Nazi Germany in World War II was long and brutal,
lasting for several years. Despite facing devastating losses, the Soviet Union kept resisting
with unyielding determination. Harsh weather conditions, especially the bitter Russian winter,
played a crucial role in slowing down the German advance. The Soviets also made effective use
of guerrilla warfare, ambushing enemy forces and disrupting supply lines. Battles like Stalingrad
turned the tide, showcasing their resilience and strategic strength. This relentless struggle helped
shape the outcome of the war."""

#(i)
cleaned_paragraph = re.sub(r'[^\w\s]', '', paragraph.lower())
print("\nCleaned Paragraph:\n", cleaned_paragraph)

#(ii)
sentences = sent_tokenize(paragraph)
word_tokens = word_tokenize(cleaned_paragraph)

print("\nSentences:")
for sent in sentences:
    print("-", sent)

print("\nWord Tokens:")
print(word_tokens)

#(iii)
split_words = cleaned_paragraph.split()
print("\nUsing split():")
print(split_words)

print("\nDifference between split() and word_tokenize():")
print("split() just splits on spaces, while word_tokenize() handles punctuation and linguistic structure more accurately.")

#(iv)
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in word_tokens if word not in stop_words]

print("\nFiltered Tokens (No Stopwords):")
print(filtered_tokens)

#(v)
fdist = FreqDist(filtered_tokens)

print("\nWord Frequency Distribution (excluding stopwords):")
for word, freq in fdist.items():
    print(f"{word}: {freq}")
```

Q2. Using the same paragraph from Q1:

1. Extract all words with only alphabets using `re.findall()`  
2. Remove stop words using NLTK’s `stopword` list  
3. Perform stemming with `PorterStemmer`  
4. Perform lemmatization with `WordNetLemmatizer`  
5. Compare the stemmed and lemmatized outputs and explain when you’d prefer one over the other.



```python
import re
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

paragraph = """Russia’s fight against Nazi Germany in World War II was long and brutal,
lasting for several years. Despite facing devastating losses, the Soviet Union kept resisting
with unyielding determination. Harsh weather conditions, especially the bitter Russian winter,
played a crucial role in slowing down the German advance. The Soviets also made effective use
of guerrilla warfare, ambushing enemy forces and disrupting supply lines. Battles like Stalingrad
turned the tide, showcasing their resilience and strategic strength. This relentless struggle helped
shape the outcome of the war."""

#(i)
words_only_alpha = re.findall(r'\b[a-zA-Z]+\b', paragraph)
print("\nWords with only alphabets:")
print(words_only_alpha)

#(ii)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_only_alpha if word.lower() not in stop_words]
print("\nFiltered words (no stopwords):")
print(filtered_words)

#(iii)
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_words]
print("\nStemmed words:")
print(stemmed_words)

#(iv)
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word.lower()) for word in filtered_words]
print("\nLemmatized words:")
print(lemmatized_words)

#(v)
print("\nComparison of stemming vs lemmatization:")
for original, stemmed, lemmatized in zip(filtered_words, stemmed_words, lemmatized_words):
    print(f"{original:<15} | Stemmed: {stemmed:<12} | Lemmatized: {lemmatized}")
```

Q3. Choose 3 short texts of your own (e.g., different news headlines, product reviews).

1. Use `CountVectorizer` to generate the Bag of Words representation.  
2. Use `TfidfVectorizer` to compute TF-IDF scores.  
3. Print and interpret the top 3 keywords from each text using TF-IDF.



```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import numpy as np

texts = [
    "The new iPhone has an amazing camera and faster processor.",
    "This laptop is lightweight, has great battery life, and a stunning display.",
    "The movie had breathtaking visuals but the storyline was weak."
]

#(i)
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(texts)
print("Bag of Words Representation:\n")
print(count_matrix.toarray())
print("\nFeature Names (BoW):")
print(count_vectorizer.get_feature_names_out())

#(ii)
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(texts)
feature_names = tfidf_vectorizer.get_feature_names_out()

print("\nTF-IDF Scores Matrix:\n")
print(np.round(tfidf_matrix.toarray(), 2))

#(iii)
print("\nTop 3 Keywords per Text (based on TF-IDF):\n")

for i, text in enumerate(texts):
    tfidf_scores = tfidf_matrix[i].toarray().flatten()
    top_indices = tfidf_scores.argsort()[-3:][::-1]
    top_words = [(feature_names[index], tfidf_scores[index]) for index in top_indices]

    print(f"Text {i+1}: \"{text}\"")
    for word, score in top_words:
        print(f"  - {word}: {score:.4f}")
    print()
```

Q4. Write 2 short texts (4–6 lines each) describing two different technologies (e.g., AI vs Blockchain).

1. Preprocess and tokenize both texts.  
2. Calculate:  
   a. Jaccard Similarity using sets  
   b. Cosine Similarity using `TfidfVectorizer` + `cosine_similarity()`  
   c. Analyze which similarity metric gives better insights in your case.



```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')
nltk.download('stopwords')

text1 = """Artificial Intelligence (AI) is a field of computer science that focuses on creating systems
that can perform tasks typically requiring human intelligence. These include learning, problem-solving,
speech recognition, and decision-making."""

text2 = """Blockchain is a decentralized digital ledger that records transactions across many computers.
It is secure, transparent, and tamper-proof, making it ideal for cryptocurrency, supply chain, and
contract management applications."""

#(i)
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    return [word for word in tokens if word not in stop_words]

tokens1 = preprocess(text1)
tokens2 = preprocess(text2)

print("Tokens from Text 1 (AI):", tokens1)
print("Tokens from Text 2 (Blockchain):", tokens2)

#(ii) a.
set1 = set(tokens1)
set2 = set(tokens2)

intersection = set1.intersection(set2)
union = set1.union(set2)
jaccard_similarity = len(intersection) / len(union)

print(f"\nJaccard Similarity: {jaccard_similarity:.4f}")

#(ii) b.
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform([text1, text2])
cosine_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]

print(f"Cosine Similarity: {cosine_sim:.4f}")

#(ii) c.
# Jaccard Similarity considers only exact word overlap (intersection/union of sets).
# Low for very different topics even if context is related.
# Cosine Similarity works on vectorized context, so it can detect semantic closeness even if word overlap is minimal.
```

Q5. Write a short review for a product or service.

1. Use `TextBlob` or `VADER` to find polarity & subjectivity for each review.  
2. Classify reviews into Positive / Negative / Neutral.  
3. Create a word cloud using the `wordcloud` library for all positive reviews.



```python
from textblob import TextBlob
!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

reviews = [
    "I recently bought the new smartphone, and it's amazing! The camera quality is fantastic, battery life is great, and the design is sleek. Highly recommend!",
    "The headphones broke within a week of use. The sound quality is poor, and the build feels cheap. Not worth the price at all.",
    "The service was okay. Nothing exceptional, but it met my expectations. I might consider it again."
]

#(i)
def textblob_analysis(review):
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity
    return polarity, subjectivity

analyzer = SentimentIntensityAnalyzer()
def vader_analysis(review):
    sentiment = analyzer.polarity_scores(review)
    return sentiment['compound']

reviews_analysis = []
for review in reviews:
    tb_polarity, tb_subjectivity = textblob_analysis(review)
    vader_polarity = vader_analysis(review)
    reviews_analysis.append({
        'review': review,
        'textblob_polarity': tb_polarity,
        'textblob_subjectivity': tb_subjectivity,
        'vader_polarity': vader_polarity
    })

#(ii)
def classify_review(polarity, threshold=0.1):
    if polarity > threshold:
        return "Positive"
    elif polarity < -threshold:
        return "Negative"
    else:
        return "Neutral"

for review in reviews_analysis:
    review['textblob_sentiment'] = classify_review(review['textblob_polarity'])
    review['vader_sentiment'] = classify_review(review['vader_polarity'])

print("\nReview Sentiment Analysis:")
for review in reviews_analysis:
    print(f"Review: {review['review']}")
    print(f"TextBlob Sentiment: {review['textblob_sentiment']}, VADER Sentiment: {review['vader_sentiment']}")
    print()

#(iii)
positive_reviews = [review['review'] for review in reviews_analysis if review['textblob_sentiment'] == "Positive"]
positive_reviews_text = " ".join(positive_reviews)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive_reviews_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Q6. Choose your own paragraph (~100 words) as training data.
1. Tokenize text using `Tokenizer()` from `keras.preprocessing.text`  
2. Create input sequences and build a simple LSTM or Dense model  
3. Train the model and generate 2–3 new lines of text starting from any seed word you provide.



```python
!pip install tensorflow
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

#(i)
text = """Artificial Intelligence (AI) is transforming industries by enabling machines to perform tasks that traditionally required human intelligence.
From speech recognition to autonomous vehicles, AI has the potential to revolutionize sectors like healthcare, finance, and entertainment.
With advancements in machine learning and deep learning, AI systems are becoming more sophisticated and capable of performing complex tasks.
While AI brings numerous benefits, it also raises ethical concerns about data privacy, job displacement, and the potential for misuse.
Despite these challenges, AI is poised to play a crucial role in shaping the future of technology and society."""

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

sequences = tokenizer.texts_to_sequences([text])[0]

#(ii)
sequence_length = 5
input_sequences = []

for i in range(sequence_length, len(sequences)):
    input_sequences.append(sequences[i-sequence_length:i+1])

input_sequences = np.array(input_sequences)

#(iii)
X, y = input_sequences[:, :-1], input_sequences[:, -1]
y = to_categorical(y, num_classes=len(tokenizer.word_index)+1)

#(iv)
model = Sequential()
model.add(Embedding(len(tokenizer.word_index)+1, 10, input_length=sequence_length))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(len(tokenizer.word_index)+1, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#(v)
model.fit(X, y, epochs=100, batch_size=32)

#(vi)
def generate_text(seed_text, next_words, model, tokenizer, max_sequence_len):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        predicted_probs = model.predict(token_list, verbose=0)
        predicted_word_index = np.argmax(predicted_probs)
        predicted_word = tokenizer.index_word[predicted_word_index]
        seed_text += " " + predicted_word
    return seed_text

seed_text = "AI"
generated_text = generate_text(seed_text, next_words=20, model=model, tokenizer=tokenizer, max_sequence_len=sequence_length)
print("Generated Text:\n", generated_text)
```
