Q1. Ask the user to enter a value (e.g., M) for a mathematical function. 
-> Generate x values from -10 to 10 using np.linspace().

```python
import numpy as np
import pandas as pd

M = int(input())
x = np.linspace(-10, 10)
def generate(x):
    return M * (x**2)

print(generate(x))
```

-> Compute y values for: 
o y= M⋅x2 
o y=M⋅sin(x) 

```python
def generate2(x):
    return M * (np.sin(x))
print(generate2(x))
```

-> Plot both functions in a single figure: 
o Use different colors and line styles. 
o Add legend, grid, and title. 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))  
plt.plot(x, generate(x), label='y = M * x^2', color='green', linestyle='-')
plt.plot(x, generate2(x), label='y = M * sin(x)', color='purple', linestyle='--')
plt.legend()
plt.grid(True)
plt.title('Plot of Two Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Q.2 Create a dataset of five subjects and scores. 
-> Convert it into a Pandas DataFrame. 

```python
data = pd.DataFrame({"Subjects": ["Cognitive Computing", "P&S", "AI", "DAA", " DBMS"],
        "Scores": [85, 92, 91, 79, 64]})

data.head()
```

-> Plot the scores using a Seaborn bar plot with: 
o Different colors for each bar. 
o Annotations on top of each bar. 
o Title, axis labels, and grid. 

```python
import seaborn as sns
plt.figure(figsize=(12, 8))
ax = sns.barplot(x="Subjects", y="Scores", data=data, palette="viridis")  
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')
plt.title("Different Subject Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and 
then generate a dataset of 50 values using np.random.randn(). 
-> Create a 2x2 subplot layout displaying: 
o Line plot showing cumulative sum. 
o Scatter plot with random noise. 

-> Customize with titles, axis labels, and grids. 

```python
rno = int(input())
np.random.seed(rno)
dataset = np.random.randn(50)
fig, axes = plt.subplots(2,2,figsize = (10,8))
axes[0, 0].plot(np.cumsum(dataset),marker ="o", linestyle="-")
axes[0, 0].set_title('Cumulative Sum')
axes[0, 0].set_xlabel('Data Point')
axes[0, 0].set_ylabel('Cumulative Sum')
axes[0, 0].grid(True)
noise = np.random.randn(50) * 0.2
axes[1, 1].scatter(dataset, dataset + noise, marker='x', color='green')
axes[1, 1].set_title('Scatter Plot with Random Noise')
axes[1, 1].set_xlabel('X-axis')
axes[1, 1].set_ylabel('Y-axis')
axes[1, 1].grid(True)
plt.show()
```

Apply ‘seaborn’ library to do the following.

```python
df = pd.read_csv("company_sales_data.csv")
df.head()
```

2. Read all product sales data and show it using a multiline plot. 

```python
fig, axes = plt.subplots(2,2,figsize = (10,8))
sns.lineplot(x="month_number", y ="total_profit",data=df, palette="viridis",ax = axes[0,0])
axes[0,0].set_title("Total_profit for all months")
axes[0,0].set_xlabel("Month number")
axes[0,0].set_ylabel("Total_profit")
axes[0,0].grid(True)
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in products:
    sns.lineplot(x='month_number', y=product, data=df, label=product, ax = axes[0,1])
plt.legend()
plt.show()
axes[0,1].set_title("Product Sales Data")
axes[0,1].set_xlabel("Month number")
axes[0,1].set_ylabel("Sales")
axes[0,1].grid(True)

plt.show()
```

3. Plot bar chart for all the features/attributes. 

```python
plt.figure(figsize=(16, 5))
df.drop(columns='month_number').sum().plot(kind='bar')
plt.title('Bar Chart for All Features/Attributes')
plt.xlabel('Features')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout() 
plt.show()
```
