Q1.  Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports,
technology, food, books, etc.).
1. Convert text to lowercase and remove punctuation.
2. Tokenize the text into words and sentences.
3. Remove stopwords (using NLTK's stopwords list).
4. Display word frequency distribution (excluding stopwords).

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

paragraph = """
Russia’s fight against Nazi Germany in World War II was long and brutal, lasting for several years.
Despite facing devastating losses, the Soviet Union kept resisting with unyielding determination.
Harsh weather conditions, especially the bitter Russian winter, played a crucial role in slowing down the German advance.
The Soviets also made effective use of guerrilla warfare, ambushing enemy forces and disrupting supply lines.
Battles like Stalingrad turned the tide, showcasing their resilience and strategic strength.
This relentless struggle helped shape the outcome of the war.
"""

paragraph = paragraph.strip()

#(i)
text = paragraph.lower()
text = text.translate(str.maketrans("", "", string.punctuation))

#(ii)
text = paragraph.lower()
sentences = sent_tokenize(text)
table = str.maketrans("","",string.punctuation)
sentences = [s.translate(table) for s in sentences]

text = paragraph.lower()
text = text.translate(str.maketrans("", "", string.punctuation))
words = word_tokenize(text)

#(iii)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

#(iv)
word_freq = Counter(filtered_words)

print("Tokenized Sentences:")
for i, sentence in enumerate(sentences, 1):
    print(f"{i}: {sentence}")

print("\nFiltered Words (excluding stopwords):")
print(filtered_words)

print("\nWord Frequency Distribution:")
for word, freq in word_freq.most_common():
    print(f"{word}: {freq}")
```

Q2: Stemming and Lemmatization on
1. Take the tokenized words from Question 1 (after stopword removal).
2. Apply stemming using NLTK's PorterStemmer and LancasterStemmer.
3. Apply lemmatization using NLTK's WordNetLemmatizer.
4. Compare and display results of both techniques.

```python
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')

#(i)
filtered_words = [
    'russia’s', 'fight', 'nazi', 'germany', 'world', 'war', 'ii', 'long', 'brutal', 'lasting', 'several', 'years',
    'despite', 'facing', 'devastating', 'losses', 'soviet', 'union', 'kept', 'resisting', 'unyielding',
    'determination', 'harsh', 'weather', 'conditions', 'especially', 'bitter', 'russian', 'winter', 'played',
    'crucial', 'role', 'slowing', 'german', 'advance', 'soviets', 'also', 'made', 'effective', 'use', 'guerrilla',
    'warfare', 'ambushing', 'enemy', 'forces', 'disrupting', 'supply', 'lines', 'battles', 'like', 'stalingrad',
    'turned', 'tide', 'showcasing', 'resilience', 'strategic', 'strength', 'relentless', 'struggle', 'helped',
    'shape', 'outcome'
]

#(ii) & (iii)
porter = PorterStemmer()
lancaster = LancasterStemmer()
lemmatizer = WordNetLemmatizer()

results = []
for word in filtered_words:
    results.append({
        "Original": word,
        "PorterStemmer": porter.stem(word),
        "LancasterStemmer": lancaster.stem(word),
        "Lemmatizer": lemmatizer.lemmatize(word)
    })

print(f"{'Original':<15}{'PorterStemmer':<20}{'LancasterStemmer':<20}{'Lemmatizer'}")
print("-" * 75)
for res in results:
    print(f"{res['Original']:<15}{res['PorterStemmer']:<20}{res['LancasterStemmer']:<20}{res['Lemmatizer']}")
```

Q3. Regular Expressions and Text Splitting
1. Take their original text from Question 1.
2. Use regular expressions to:
a. Extract all words with more than 5 letters.
b. Extract all numbers (if any exist in their text).
c. Extract all capitalized words.
3. Use text splitting techniques to:
a. Split the text into words containing only alphabets (removing digits and special
characters).
b. Extract words starting with a vowel.

```python
import re

#(i)
text = """
Russia’s fight against Nazi Germany in World War II was long and brutal, lasting for several years.
Despite facing devastating losses, the Soviet Union kept resisting with unyielding determination.
Harsh weather conditions, especially the bitter Russian winter, played a crucial role in slowing down the German advance.
The Soviets also made effective use of guerrilla warfare, ambushing enemy forces and disrupting supply lines.
Battles like Stalingrad turned the tide, showcasing their resilience and strategic strength.
This relentless struggle helped shape the outcome of the war.
"""

# (ii) a.
words_over_5 = re.findall(r'\b\w{6,}\b', text)

# (ii) b.
numbers = re.findall(r'\b\d+\b', text)

# (ii) c.
capitalized_words = re.findall(r'\b[A-Z][a-z]*\b', text)

# (iii) a.
alphabetic_words = re.findall(r'\b[a-zA-Z]+\b', text)

# (iii) b.
vowel_words = re.findall(r'\b[aeiouAEIOU][a-zA-Z]*\b', text)

print("Words with more than 5 letters:\n", words_over_5)
print("\nNumbers in the text:\n", numbers)
print("\nCapitalized words:\n", capitalized_words)
print("\nWords with only alphabets:\n", alphabetic_words)
print("\nWords starting with a vowel:\n", vowel_words)
```
